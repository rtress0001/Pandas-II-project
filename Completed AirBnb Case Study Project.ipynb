```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019 - AB_NYC_2019.csv' sep = ",")
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?

print("The number of unique nieghbourhood groups in the data is :",air_bnb['neighbourhood_group'].nunique())
air_bnb.groupby('neighbourhood_group').count()
print("Manhattan has the most")
```

```python
# Are private rooms the most popular in manhattan?
manhattan_rooms = air_bnb.query("neighbourhood_group == 'Manhattan'")
manhattan_rooms.groupby('room_type').count()
print("Private rooms are not the most popular in manhattan")

```

```python
# Which hosts are the busiest and based on their reviews?
sum_host_name_reviews = air_bnb.groupby('host_name')['reviews_per_month'].sum()
busy_hosts =sum_host_name_reviews.sort_values(ascending=False)
print(busy_hosts.head(1))

sum_host_name_reviews2 = air_bnb.groupby('host_name')['number_of_reviews'].sum()
busy_hosts2 =sum_host_name_reviews2.sort_values(ascending=False)
print(busy_hosts2.head(1))

print("David has the highest velocity of reviews, but Micheal has the highest number of reviews")

```

```python
#Which neighorhood group has the highest average price?
answer = air_bnb.groupby('neighbourhood_group')['price'].mean()
#find the index of nieghbourhood group
index_group = answer.idxmax()
print(f'The group {index_group} has the price of {answer.max()}')
```

```python
# Which neighbor hood group has the highest total price?
answer2 = air_bnb.groupby('neighbourhood_group')['price'].sum()
#find the index of nieghbourhood group
index_group2 = answer.idxmax()
print(f'The group {index_group2} has the price of {answer2.sum()}')
```

```python
#Which top 5 hosts have the highest total price?
sum_host_name_price = air_bnb.groupby('host_name')['price'].sum()
answer_sum = sum_host_name_price.sort_values(ascending=False)
print(answer_sum.head(5))

```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
availabity_review_count_filt = air_bnb.query('number_of_reviews > 100 and availability_365 == 0')
count=availabity_review_count_filt.shape[0]
print(f'The Number of reviews that meet these conditions are {count}')

```

```python
# What host has the highest total of prices and where are they located?
answer_high_total_by_host_name = air_bnb.groupby('host_name')['price'].sum()
print(air_bnb.query('host_name == "Sonder (NYC)"'))
index_group_total_price = answer_high_total_by_host_name.idxmax()
print(index_group_total_price)
neighbourhood_group_counts = air_bnb.query('host_name == "Sonder (NYC)"')['neighbourhood_group'].value_counts()
print(f'The individual of {index_group_total_price} has listings in  {neighbourhood_group_counts}')


```
